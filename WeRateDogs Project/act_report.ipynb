### Virtual Analysis Report

This goal of this project is to critically look into a twitter user's page called WeRateDogs, where we plan to help analyze the users page in answering questions such as;

- What dog stage was most prominent across all images posted on the page
- User behaviour of the followers of the page across the time captured (declined/Improved)
- Most popular dogs that appeared in the page
- Relationships between followers, retweets and favourites on the page and what the influence could be.




#### Data Gathering 
Three datasets were considered for this project and merged to get a master dataset used in further analysis and insight generation

- WeRateDogs twitter archive data that contains all basic tweet information
- Image prediction file that contains predictions of the dogs posted on the page
- Tweet info data that contains additional information on the tweets such as retweet count, favourites etc
